```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

Loading the data 

```python
df = pd.read_csv('dataset.csv')
df.head()
```

Loading the test data

Data cleaning: Missing values

```python
df.fillna(value=df['speed'].mean(), inplace=True)
```

```python
df.info()
```

Data normalization

```python
scaler = MinMaxScaler()
scaler.fit(df[['speed']])
df['speed'] = scaler.transform(df[['speed']])
df.head()
```

```python
df['speed'].mean()
```

PCA

```python
pca = PCA(n_components=1) 

df['weight'] = pca.fit_transform(df[['minute', 'is_holiday', 'weather', 'length']])
```

```python
df
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(df['length'], df['speed'], s=0.01, c='black')
plt.xlabel('length')
plt.ylabel('speed')
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(df['length'], df['minute'], s=0.1, c='black')
plt.xlabel('length')
plt.ylabel('minute')
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(df['speed'], df['minute'], s=0.1, c='black')
plt.xlabel('speed')
plt.ylabel('minute')
```

```python
def histogram(column):
    plt.figure(figsize=(8, 5))
    plt.hist(df[column], bins=300)
    plt.xlabel('edges')
    plt.ylabel('length')
    plt.title(column)
    plt.show()
```

```python
histogram('length')
```
